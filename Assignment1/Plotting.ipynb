```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("B7_PlotAPg.csv")

# Create a wider figure
plt.figure(figsize=(12, 6))  # Wider aspect ratio for better visibility of the data

# Plot for each generator
for gen in df["Generator"].unique():
    gen_data = df[df["Generator"] == gen]
    plt.plot(gen_data["CO2_Limit"], gen_data["AP_g"], 
             linestyle="-",  # Lines only, no markers
             label=gen, 
             linewidth=2)    # Thicker lines for better visibility

# Formatting
plt.xlabel("CO₂ Limit (Tonnes per hour)")
plt.ylabel("Average Power Output (MW)")
plt.title("Average Power Output of each generator as functions of the CO2 limit between the values in B4 and B6+500t/h")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Load the CSV file
df = pd.read_csv("B7_PlotAC.csv")

# Sort the DataFrame based on CO2_Limit to ensure the data is in order
df = df.sort_values(by="CO2_Limit")

# Plot with adjusted size and line for visualizing each section
plt.figure(figsize=(12, 6))  # Increase the width for more horizontal space
plt.plot(df["CO2_Limit"], df["AC"], linestyle="-", color="g", linewidth=2)  # Change color to green

# Formatting
plt.xlabel("CO₂ Limit (Tonnes per hour)")
plt.ylabel("Average Hourly Cost in £")
plt.title("Average hourly cost (£) of each generator as a function of the CO2 limit between the values in B4 and B6+500t/h")

# Optionally adjust the x and y axis limits to make the plot more spread out
plt.xlim(min(df["CO2_Limit"]) - 50, max(df["CO2_Limit"]) + 50)  # Padding for x-axis
plt.ylim(min(df["AC"]) - 500, max(df["AC"]) + 500)  # Padding for y-axis

# Set the y-axis to display in regular numeric format, not scientific notation
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

plt.grid(True)

# Show the plot
plt.show()
```
